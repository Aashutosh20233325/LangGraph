```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class BatsMan(TypedDict):
    name: str
    runs: int
    balls: int
    fours: int
    sixes: int
    strike_rate: float
    bpb: float  # balls per boundary
    boundary_percentage: float
    summary: str
```

```python
def calculate_strike_rate(state: BatsMan) -> BatsMan:
    sr = (state['runs'] / state['balls']) * 100 if state['balls'] > 0 else 0.0
    return {'strike_rate': sr}
def calculate_bpb(state: BatsMan) -> BatsMan:
    bpb = state['balls'] / (state['fours'] + state['sixes']) if (state['fours'] + state['sixes']) > 0 else float('inf')
    return {'bpb': bpb}
def calculate_boundary_percentage(state: BatsMan) -> BatsMan:
    total_boundaries = state['fours'] + state['sixes']
    boundary_percentage = (total_boundaries / state['balls']) * 100 if state['balls'] > 0 else 0.0
    return {'boundary_percentage': boundary_percentage} 
def generate_summary(state: BatsMan) -> BatsMan:
    summary = (f"{state['name']} scored {state['runs']} runs off {state['balls']} balls, "
               f"with a strike rate of {state['strike_rate']:.2f}, "
               f"including {state['fours']} fours and {state['sixes']} sixes. "
               f"Balls per boundary: {state['bpb']:.2f}, "
               f"Boundary percentage: {state['boundary_percentage']:.2f}%.")
    print(summary)
    return {'summary': summary}
```

```python
graph = StateGraph(BatsMan)
graph.add_node('caculate_sr',calculate_strike_rate)
graph.add_node('calculate_bpb',calculate_bpb)
graph.add_node('calcuulate_boundary_percent',calculate_boundary_percentage)
graph.add_node('summary',generate_summary)


graph.add_edge(START,'caculate_sr')
graph.add_edge(START,'calculate_bpb')
graph.add_edge(START,'calcuulate_boundary_percent')

graph.add_edge('caculate_sr','summary')
graph.add_edge('calculate_bpb','summary')
graph.add_edge('calcuulate_boundary_percent','summary')

graph.add_edge('summary',END)

workflow = graph.compile()
```

```python
intial_state = {
    'name': 'John Doe',
    'runs': 85,
    'balls': 60,
    'fours': 8,
    'sixes': 2}
workflow.invoke(intial_state)
```
