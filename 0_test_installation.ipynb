```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
#define state
class BMIState(TypedDict):
    weight_kg:float
    height_m:float
    bmi:float
    category:str
    
```

```python
def calculate_bmi(state:BMIState) ->BMIState:#WHEN NODE IS EXECUTED IT IS GIVEN #STATE OF GRAPH AND IT RETUEND MDIFED STATE ITSELT
    bmi = state['weight_kg'] / (state['height_m'] ** 2)
    state['bmi'] = round(bmi, 2)
    return state

def label_bmi(state:BMIState) ->BMIState:
    bmi = state['bmi']
    if bmi < 18.5:
        label = 'Underweight'
    elif 18.5 <= bmi < 24.9:
        label = 'Normal weight'
    elif 25 <= bmi < 29.9:
        label = 'Overweight'
    else:
        label = 'Obesity'
    state['category'] = label
    return state
```

```python
#define graph
graph = StateGraph(BMIState)

#add nodes to your graph
graph.add_node('calculate_bmi(name_of_node)',calculate_bmi)
graph.add_node('label_bmi(name_of_node)',label_bmi)
#add edges to graph
graph.add_edge(START,'calculate_bmi(name_of_node)')
graph.add_edge('calculate_bmi(name_of_node)','label_bmi(name_of_node)')
graph.add_edge('label_bmi(name_of_node)',END)


#compile the graph
workflow = graph.compile()
```

```python

#execute the graph
initial_state = {'weight_kg':70,'height_m':1.75}
final_state = workflow.invoke(initial_state)
print(final_state)
```
