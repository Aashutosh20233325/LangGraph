```python
from langchain_deepseek import ChatDeepSeek
from langchain_core.messages import HumanMessage, SystemMessage
import os
os.environ["DEEPSEEK_API_KEY"] = "sk-8bdf3195aabd4d89ade18d19e3340438"
# Initialize the DeepSeek chat model
# Common models:
# - "deepseek-chat"      : General chat (DeepSeek-V3, fast and versatile)
# - "deepseek-reasoner" : Reasoning mode (DeepSeek-R1, shows thought process)
# - "deepseek-coder"     : If available for coding tasks
model = ChatDeepSeek(
    model="deepseek-chat",  # Change this to test other models
    temperature=0.7,        # Adjust for creativity (0 = deterministic)
    max_tokens=1024,
)

print("DeepSeek Chatbot ready! Type 'quit' to exit.\n")

# Optional: Add a system prompt for personality
system_prompt = SystemMessage(content="You are a helpful and friendly AI assistant.")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["quit", "exit", "bye"]:
        print("Goodbye!")
        break
    
    # Build messages (include system prompt if desired)
    messages = [system_prompt, HumanMessage(content=user_input)]
    
    # Stream the response for a more interactive feel
    print("Assistant: ", end="", flush=True)
    for chunk in model.stream(messages):
        print(chunk.content, end="", flush=True)
    print("\n")
```
